# DATA VISUALIZATION AND DATA ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the datasets
data=pd.read_csv('Parkinsons.csv')
data.head()
```

```python
data.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

var = ['Symptom', 'Alzheimer_Disease', 'Parkinsons_Disease',
       'Overlapping_Symptom']
  
for i in var:
    data[i] = le.fit_transform(data[i]).astype(int)
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x='Parkinsons_Disease',data=data)
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.hist(data['Alzheimer_Disease'],color='blue')

plt.subplot(1,2,2)
plt.hist(data['Parkinsons_Disease'],color='green')
```

```python
data.hist(figsize=(15,60),layout=(15,4),color='green')
plt.show()
```

```python
plt.boxplot(data) 
```

```python
sns.pairplot(data)
```

```python
fig,ax=plt.subplots(figsize=(20,15))
sns.heatmap(data.corr(),annot=True,fmt='0.2%',cmap='autumn',ax=ax)
```

```python
def plot(df, variable):
    dataframe_pie = data[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(9,9), autopct='%1.2f%%', fontsize = 10)
    ax.set_title(variable + ' \n', fontsize = 10)
    return np.round(dataframe_pie/df.shape[0]*100,2)

plot(data, 'stroke')
```
