```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

var = ['Symptom', 'Alzheimer_Disease', 'Parkinsons_Disease',
       'Overlapping_Symptom']
  
for i in var:
    df[i] = le.fit_transform(df[i]).astype(int)
```
